```python
df['total marks']=df['math score']+df['writing score']+df['reading score']#finding total marks
df['marks']=df['total marks']/3#finding percentage==marks
df.head()
```

```python
#ASSIGNING GRADES
def finding_grades(marks):
    if marks>85 and marks<=100:
        return 'Grade A'
    elif marks>70 and marks<=85:
        return 'Grade B'
    elif marks>55 and marks<=70:
        return 'Grade C'
    elif marks>40 and marks<=55:
        return 'Grade D'
    elif marks>=0 and marks<=40:
        return 'Grade E'

df['grades']=df['marks'].apply(finding_grades)
```

```python
df.head()
```

```python
df.info()
```

```python
df['grades'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

**Most of the students obtained grade B and C

**Very few students obtained grade E

**11.2% of students obtained grade A

```python
plt.title('GENDER VS GRADES')
sns.countplot(hue='gender',x='grades',data=df)
plt.show()
plt.title('RACE VS GRADES')
sns.countplot(x='grades',hue='race/ethnicity',data=df)
plt.show()
plt.title('GRADES VS LUNCH')
sns.countplot(hue='grades',x='lunch',data=df)
plt.show()
plt.title('TEST PREPARATION COURSE VS GRADES')
sns.countplot(hue='test preparation course',x='grades',data=df)
plt.show()
plt.title('GRADES VS PARENTAL LEVEL OF EDUCATION')
sns.countplot(hue='parental level of education',x='grades',data=df)
plt.show()
```

**Grade A obtained by female students is more compared to the male students

**Grade E is obtained by female and male students is almost same

**Students having standard lunch obtained nice grades compared to students having reduced/free lunch

**In all the groups high proportion of students obtained grades C and D

**Almost negligible number of students obtained grade E who had completed their test preparation course

**Grade A obtained by students who has completed their test preperation course is more compared to students who has done nothing

**For students whose parental level of education is "MASTER DEGREE",the probability of getting grade E is 0
